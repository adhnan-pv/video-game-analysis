```python
vg6=vg[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].groupby('Genre').sum()
```

```python
vg6
```

```python

plt.plot(vg6['Genre'],vg6['Global_Sales'],label='Global_Sales',color='r',marker='<',markersize=6,linewidth=3)
plt.plot(vg6['Genre'],vg6['NA_Sales'],label='north america',color='g',marker='^',markersize=6,linewidth=3)
plt.plot(vg6['Genre'],vg6['EU_Sales'],label='Europe',color='k',marker='P',markersize=6,linewidth=3)
plt.plot(vg6['Genre'],vg6['JP_Sales'],label='japan',color='b',marker='p',markersize=6,linewidth=3)
plt.plot(vg6['Genre'],vg6['Other_Sales'],label='other',color='m',marker='o',markersize=6,linewidth=3)


plt.xlabel('Genre')
plt.ylabel('sales (M)')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
